# Fourier transform approach


```python
import numpy as np
import seaborn as sn
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fftpack import fft
import os
import json
from sklearn.model_selection import train_test_split
```

```python
# Number of frequencies to use, defined in the json file
N = 10
JSONdata = {}
count = 0
```

```python
print("OPENING FILE")
# Adding datas to JSONdata
with open('data_FULL_Version_2.json') as json_file:
    JSONdata['sounds'] = json.load(json_file)['sounds']
print("END FILE")
```

```python
# Adding data from JSON to correct arrays to use later with scikit learn
Y = np.array([])
X = np.array(np.empty(shape=[0, N]))
print(X.shape)
print("DEBUT AJOUT DANS ARRAY")
for arr in JSONdata['sounds']:
    a = np.array(arr['maxs'])
    X = np.concatenate((X,[a]))
    Y = np.append(Y,arr['label'])
    count += 1
    if count%100 == 0:
        print(count)
    if count == 50000:
        break
print("FIN AJOUT DANS ARRAY")
```

```python
# Spliting datas into two sets, one for training, and one for testing
Xtrain, Xtest, ytrain, ytest = train_test_split(X, Y, test_size= 0.2, random_state=0)
print(Xtrain.shape)
print(Xtest.shape)
print('pourcentage:' ,Xtrain.shape[0]/X.shape[0])
```

```python
print("DEBUT TRAIN")
from sklearn.tree import DecisionTreeClassifier
# Creation of the DecisionTreeClassifier
Arbre_decision = DecisionTreeClassifier(random_state=0, max_depth=20)
# Fitting data
clf = Arbre_decision.fit(Xtrain, ytrain)
print("FIN TRAIN")
from sklearn.metrics import accuracy_score
# Predicting our test set
ypredit = clf.predict(Xtest)
print(accuracy_score(ytest, ypredit))
```

```python
from sklearn import metrics
# The following variable contains labels that will be used for plotting the confusion matrix
theLabels=['bass','brass','flute','guitar','keyboard','mallet','organ','reed','string','synth','vocal']
#Creating the confusion matrix
array = (metrics.confusion_matrix(ytest, ypredit,labels=theLabels))
# Creating the DataFrame
leDataFrame = pd.DataFrame(array, index = theLabels,columns = theLabels)
plt.figure(figsize = (10,10))

# Plotting with seaborn
ax = sn.heatmap(leDataFrame, annot=True,cmap="Blues",square=True,fmt='d')
ax.set_ylim(len(array), -0.5)
```
